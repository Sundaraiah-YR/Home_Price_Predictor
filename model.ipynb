```python
df=pd.read_csv('Bengaluru_House_Data.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# dropping all the unwanted columns
df=df.drop(["area_type","availability","society","balcony"],axis=1)
df.info()
```

```python
# Handling missing values in Location column
df['location'].value_counts() # as whitefield is repeated for more times, fill with the same
df.info()
```

```python
df['location']=df['location'].fillna('Whitefield')
df.info()
```

```python
# handling missing values in size columns
df['size'].value_counts()   # as 2 BHK is repeated for more times, fill with the same
```

```python
df['size']=df['size'].fillna('2 BHK')
df.info()
```

```python
# handling missing values in bath column
df['bath']=df['bath'].fillna(df['bath'].median())
df.info()
```

```python
df.head()
```

```python
# cleaning size column
df['size'].unique()
out=[i.split()[0] for i in df['size']]  # we can use int here , for integer like int(i.split()[0])
df['bhk']=out
df.head()
```

```python
df.info()
```

```python
# convert bhk column from object to int datatype
df['bhk']=df['bhk'].astype('int')
df.info()
```

```python
# cleaning total_sqft
df['total_sqft'].unique()
def convert(a):
    l=a.split('-')
    if len(l)==2:
        return (float(l[0])+float(l[1]))/2
    try:
        return float(a)
    except:
        return None
df['total_sqft']=df['total_sqft'].apply(convert)
df.info()
```

```python
df['total_sqft']=df['total_sqft'].fillna(df['total_sqft'].median()) # i have doubt 1 bhk means we cant have 2000sqft median value
df.info()
```

```python
df['price_per_sqft']=df['price']*100000/df['total_sqft']
df['price_per_sqft']
```

```python
df.head()
```

```python
# cleaning location column
df['location']=df['location'].apply(lambda x:x.strip())
df['location'].value_counts()
```

```python
location_count=df['location'].value_counts()
loc_less_than_10=location_count[location_count<=10]
loc_less_than_10
```

```python
df['location']=df['location'].apply(lambda x:'Other' if x in loc_less_than_10 else x)
df['location'].value_counts()
```

```python
# Handling the outliers in perice_per_sqft using IQR method
q1=df['price_per_sqft'].quantile(0.25)
q3=df['price_per_sqft'].quantile(0.75)
IQR=q3-q1
low_bound=q1-0.5*IQR
upp_bound=q3+0.5*IQR
df = df[(df['price_per_sqft'] >= low_bound) & (df['price_per_sqft'] <= upp_bound)]
sns.boxplot(y='price_per_sqft',data=df)
```

```python
copied_dataset=df.copy()  # shallow copy
```

```python
from sklearn.preprocessing import LabelEncoder
obj=LabelEncoder()
df['location']=obj.fit_transform(df['location'])
df.head()
```

```python
df.reset_index(inplace=True)
df=df.drop('index',axis=1)
df
```

```python
copied_dataset
```

```python
X=df.drop('price',axis=1)
y=df['price']
```

```python
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
model=RandomForestRegressor(random_state=42)
params={
    "n_estimators":[100,150,200,250,300],
    "max_depth":[3,4,5,6]
}
```

```python
grid=GridSearchCV(estimator=model,param_grid=params,cv=5)
grid.fit(Xtrain,ytrain)
```

```python
print(f'training efficiency:{grid.score(Xtrain,ytrain)}')
print(f'testinge fficiency:{grid.score(Xtest,ytest)}')
```

```python
with open('RFmodel.pkl','wb') as file:
    pickle.dump(grid,file)
```
